```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
```

```python
print("Upload an image:")
uploaded = files.upload()
image_path = list(uploaded.keys())[0]
```

```python
image = cv2.imread(image_path)
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
gray_float = np.float32(gray_image)
```

```python
harris_corners = cv2.cornerHarris(gray_float, blockSize=2, ksize=3, k=0.04)
```

```python
harris_corners_dilated = cv2.dilate(harris_corners, None)
```

```python
corner_threshold = 0.01 * harris_corners.max()
image_with_corners = image.copy()
image_with_corners[harris_corners > corner_threshold] = [0, 0, 255]
```

```python
plt.figure(figsize=(15, 5))
titles = ['Original Image', 'Harris Corners']
images = [image, image_with_corners]

for i, (img, title) in enumerate(zip(images, titles)):
    plt.subplot(1, len(images), i + 1)
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.axis('off')
plt.tight_layout()
plt.show()
output_file = "image_with_harris_corners.jpg"
cv2.imwrite(output_file, image_with_corners)
```
